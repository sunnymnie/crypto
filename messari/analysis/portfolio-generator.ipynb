```python
portfolio = portfolio.split("\n\n")
```

```python
portfolio = list(map(lambda x: x.replace("\n", ""), portfolio))
```

```python
portfolio = list(map(lambda x: x[x.find(":")+1:], portfolio))
```

```python
portfolio = list(map(lambda x: x.replace(" ", ""), portfolio))
portfolio = list(map(lambda x: x.split(","), portfolio))
```

```python
portfolio = [j for i in portfolio for j in i]
```

```python
portfolio = dict(Counter(portfolio))
```

```python
portfolio = {k: v for k, v in sorted(portfolio.items(), key=lambda item: item[1], reverse=True)}
```

## Filter to minimum 2 people holding that asset

```python
fportfolio = {key: value for (key, value) in portfolio.items() if value > 1} 
```

```python
fig, ax = plt.subplots(2, figsize=(15, 16), gridspec_kw={'height_ratios': [1, 3]})
ax[0].bar(fportfolio.keys(), fportfolio.values())
ax[0].set_xlabel("Asset")
ax[0].set_ylabel("Number of employees")
ax[1].pie(fportfolio.values(), labels=fportfolio.keys(), autopct='%1.2f%%');
```
